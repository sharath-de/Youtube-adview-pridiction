## Simple Linear Regression

```python
"""Simple Linear Regression"""

#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Loading the dataset
dataset = pd.read_csv('Salary_Data-1.csv').values

#Specify independent and dependent veraibles
x = dataset[:, :-1]
y = dataset[:, -1]
x,y
```

```python
#Visualising the data before regression
plt.scatter(x, y)
```

```python
"""Fitting Simple Linear Regression"""

#Importing linear regreassion class from sklearn library
from sklearn.linear_model import LinearRegression

#1 Create linear regression object using the imported class
model = LinearRegression()

#2 Train the model by applying 'fit fucntion
model.fit(x, y)

#Predicting the outcome by applying 'predict fucntion on input
y_pred = model.predict(x)
y_pred
```

```python
#Visualising the outcome
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')

model.intercept_
model.coef_
#
```

```python
#Visualising the outcome
plt.scatter(x, y)
plt.plot(x, y_pred, color='green')

model.intercept_
model.coef_
#
```

## Multivariable Linear Regression Model

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = [[0, 1], [5, 1], [15, 2], [25, 5], [35, 11], [45, 15], [55, 34], [60, 35]]
y = [4, 5, 20, 14, 32, 22, 38, 43]
x, y = np.array(x), np.array(y)
```

```python
 print(x)
 print(y)

 #In multiple linear regression, x is a two-dimensional array with at least two columns, 
 #while y is usually a one-dimensional array. This is a simple example of multiple linear regression,
 # and x has exactly two columns.
```

```python
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)

#You obtain the value of 𝑅² using .score() and the values of the estimators of regression coefficients
# with .intercept_ and .coef_. Again, .intercept_ holds the bias 𝑏₀, 
#while now .coef_ is an array containing 𝑏₁ and 𝑏₂ respectively.

#In this example, the intercept is approximately 5.52, 
#and this is the value of the predicted response when 𝑥₁ = 𝑥₂ = 0.
# The increase of 𝑥₁ by 1 yields the rise of the predicted response by 0.45. Similarly, when 𝑥₂ grows by 1, 
#the response rises by 0.26.
```

```python
#Predictions also work the same way as in the case of simple linear regression:
y_pred = model.predict(x)
print('predicted response:', y_pred, sep='\n')
```

```python
#The predicted response is obtained with .predict(), which is very similar to the following:
y_pred = model.intercept_ + np.sum(model.coef_ * x, axis=1)
print('predicted response:', y_pred, sep='\n')
```

```python
#You can apply this model to new data as well
x_new = np.arange(10).reshape((-1, 2))
print(x_new)

y_new = model.predict(x_new)
print(y_new)
```

```python
X = dataset[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates','alcohol']].values
y = dataset['quality'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
"""Simple Linear Regression"""

#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Loading the dataset
dataset = pd.read_csv('House_MLR.csv').values

#Specify independent and dependent veraibles
x = dataset[:, :-1]
y = dataset[:, -1]
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size = 0.1)

#fir simple linear model to training test
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
#Predicting test set
y_pred = model.predict(x_test)

#accuracy score
model.score(x,y)
```

```python
model.coef_ , model.intercept_
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test,y_pred)
r2_score
```
